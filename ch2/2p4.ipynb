6. plot the funtion $y = f(x) = x^3 - \frac{1}{x}$ and plot its tangent line at $x=1$.

```python
import numpy as np
from d2l import torch as d2l

def f(x):
    return x**3 - 1/x

# the derivative of f(x) is f'(x) = 3x^2 + 1/x^2,  and its value at x=1 is 2.
# we can derive the tangent line of f(x) at x=1 is y = 4*x - 4.

x = np.arange(0.1, 2.1, 0.1)
d2l.plot(x, [f(x), 4*x - 4], 'x', 'f(x)', legend=['f(x)', 'Tangent line(x=1)'])
```
